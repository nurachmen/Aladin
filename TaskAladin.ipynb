### Function Read Json

```python
def oke(tabel,data):
    def percobaan():
        with open(tabel,'r') as f:
            df = json.loads(f.read())
            for i in df :
                if i == 'data':
                    dataframe = pd.json_normalize(df['data'])
                    return dataframe
                elif i == 'set':
                    da = pd.json_normalize(df['set'])
                    a = ' '.join(map(str, list(set(data.columns.tolist()) & set(da.columns.tolist()))))
                    if a == '':
                        data[' '.join(map(str, list(da.columns.tolist())))] \
                        = da[' '.join(map(str, list(da.columns.tolist())))].iloc[0]
                    elif len(a.split()) == 2 :
                        a = a.split()
                        data[a[0]] = da[a[0]].iloc[0]
                        data[a[1]] = da[a[1]].iloc[0]
                    else:
                        data[a] = da[a].iloc[0]
                    return data
    return percobaan()
```
